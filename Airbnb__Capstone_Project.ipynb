### ** Treating Outliers**



```python
# figsize
plt.figure(figsize=(15,5))
# doing boxplot of numerical features
sns.boxplot(data=airbnb[numerical_features])
plt.xlabel('numerical features')
plt.ylabel('numerical values')
plt.title('Boxplot of numerical features')
plt.show()
```

- In this methos we cap an outliers in our data that means if the value is higher than ot lower than a certain thresold, all values willn be considered outliers. the metod replace the values that fall outside the specifies range with either the minimum ot maximum value within the range.

##We are going to replace the datapoints with outliers with upper and lower bound of all outliers

```python
def clip_outliers(airbnb):
  for col in airbnb[numerical_features]:
    q1 = airbnb[col].quantile(0.25)# used iqr method to define range of upper and lower outliers
    q2 = airbnb[col].quantile(0.75)
    iqr = q2-q1
    lower_bound = q1 - (1.5*iqr)
    upper_bound = q2 + (1.5*iqr)

    # replacing the outliers with upper and lower outliers
    airbnb[col] = airbnb[col].clip(lower_bound,upper_bound)
  return airbnb
```

```python
#using the funtion to treat outliers
airbnb = clip_outliers(airbnb)
```

```python
# checking the boxplot after treating outliers
plt.figure(figsize=(15,5))
sns.boxplot(data=airbnb[numerical_features])
plt.xlabel('numerical features')
plt.ylabel('numerical values')
plt.title('Boxplot of numerical features')
plt.show()
```

- Checking distribution after treating outliers

```python
for col in numerical_features:
  fig,ax =  plt.subplots(1,2,figsize=(15,3))
  sns.distplot(airbnb[col],ax=ax[0]).set(title ='distribution ')
  sns.boxplot(airbnb[col],ax=ax[1]).set(title = 'boxpot')
  plt.suptitle(f'{col.title()}', fontweight='bold')
  plt.show()


```
